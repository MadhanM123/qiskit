```python
#Imports
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(3, 3) #3 qubits, 3 classic bits

#Gates used to encode number 6
qc.x([1, 2]) #X gates on 1 and 2
qc.draw("mpl")
```

```python
qc.measure([0, 1, 2], [0, 1, 2]) #Measure qubits 0,1,2 to classic bits 0,1,2,
qc.draw("mpl")
```

```python
#Run circuit on Aer Simulator
sim = AerSimulator()
job = sim.run(qc)

result = job.result()
info = result.get_counts()

print(info)
```
