```python
#Imports
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
#Adds 1 + 1

qc = QuantumCircuit(4, 2) #2 qubits, 2 classic bits
qc.x([0, 1]) #Encode gate 0 and 1 as 1

qc.cx(0, 2) #CNOT from 0 to 2
qc.cx(1, 2) #CNOT from 1 to 2

qc.ccx(0, 1, 3) #Toffioli from 0 to 1 to 3

qc.draw("mpl")
```

```python
qc.measure(2, 0)
qc.measure(3, 1)
qc.draw("mpl")
```

```python
sim = AerSimulator()

job = sim.run(qc)

result = job.result()
info = result.get_counts()

print(info)
```
